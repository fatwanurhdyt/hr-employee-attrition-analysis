# **Exploratory Data Analysis**

## **Attrition Distribution**

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# bar chart
sns.countplot(data=hr, x='Attrition', palette='pastel', ax=axes[0], order=['No', 'Yes'])
axes[0].set_title('Number of Employees by Attrition', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Attrition', fontsize=10)
axes[0].set_ylabel('Number of Employees', fontsize=10)

for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height())}', 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='bottom', fontsize=10, xytext=(0, 5), textcoords='offset points')

# pie chart
attrition_counts = hr['Attrition'].value_counts()
axes[1].pie(attrition_counts, 
            labels=attrition_counts.index, 
            autopct='%1.1f%%',       
            startangle=90,          
            colors=sns.color_palette('pastel'), 
            explode=[0, 0.1])       
axes[1].set_title('Attrition Rate (%)', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

## **Attrition by Department**

```python
# 1. list all the categorical columns
categorical_cols = [
    'Department', 'JobRole', 'OverTime', 
    'BusinessTravel', 'Gender', 'MaritalStatus', 'EducationField'
]

# 2. change 'Attrition' column to numeric for easier calculation of attrition rate
hr['Attrition_Num'] = hr['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# 3. loop through each categorical column to create the two graphs
for col in categorical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(16, 5))
    
    # countplot (number of employees))
    sns.countplot(data=hr, x=col, hue='Attrition', palette='Set2', ax=axes[0])
    axes[0].set_title(f'Number of Employees Based on {col}', fontsize=12, fontweight='bold')
    axes[0].set_ylabel('Count')
    axes[0].tick_params(axis='x', rotation=45) 
    
    # bar chart (attrition rate)
    attr_rate = hr.groupby(col)['Attrition_Num'].mean().sort_values(ascending=False) * 100
    
    sns.barplot(x=attr_rate.index, y=attr_rate.values, palette='Reds_r', ax=axes[1])
    axes[1].set_title(f'Attrition Rate Based on {col} (%)', fontsize=12, fontweight='bold')
    axes[1].set_ylabel('Attrition Percentage (%)')
    axes[1].tick_params(axis='x', rotation=45)
    
    for p in axes[1].patches:
        axes[1].annotate(f'{p.get_height():.1f}%', 
                         (p.get_x() + p.get_width() / 2., p.get_height()), 
                         ha='center', va='bottom', fontsize=10, xytext=(0, 5), textcoords='offset points')
    
    plt.tight_layout()
    plt.show()

hr.drop(columns=['Attrition_Num'], inplace=True)
```

## **Attrition by Satisfaction Factors**

```python
# 1. satisfaction columns
satisfaction_cols = [
    'EnvironmentSatisfaction', 'JobSatisfaction', 
    'RelationshipSatisfaction', 'JobInvolvement', 'WorkLifeBalance'
]

# 2. change 'Attrition' column to numeric for easier calculation of attrition rate
if 'Attrition_Num' not in hr.columns:
    hr['Attrition_Num'] = hr['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# a. visualization loop for satisfaction columns
for col in satisfaction_cols:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # countplot
    sns.countplot(data=hr, x=col, hue='Attrition', palette='Set2', ax=axes[0])
    axes[0].set_title(f'{col} Distribution', fontsize=12, fontweight='bold')
    axes[0].set_ylabel('Number of Employees')
    axes[0].set_xlabel('Score (1=Low, 4=High)')
    
    # attrition rate
    attr_rate = hr.groupby(col)['Attrition_Num'].mean() * 100
    
    # color palette
    sns.barplot(x=attr_rate.index, y=attr_rate.values, palette='Reds_r', ax=axes[1])
    axes[1].set_title(f'Attrition Rate vs {col} (%)', fontsize=12, fontweight='bold')
    axes[1].set_ylabel('Attrition Percentage (%)')
    axes[1].set_xlabel('Score (1=Low, 4=High)')
    
    # label
    for p in axes[1].patches:
        axes[1].annotate(f'{p.get_height():.1f}%', 
                         (p.get_x() + p.get_width() / 2., p.get_height()), 
                         ha='center', va='bottom', fontsize=10, xytext=(0, 5), textcoords='offset points')
    
    plt.tight_layout()
    plt.show()

# b. heatmap correlation
plt.figure(figsize=(6, 5))
corr_matrix = hr[satisfaction_cols + ['Attrition_Num']].corr()
korelasi_target = corr_matrix[['Attrition_Num']].drop('Attrition_Num').sort_values(by='Attrition_Num', ascending=False)
sns.heatmap(korelasi_target, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation of Satisfaction Factors with Attrition', fontweight='bold')
plt.show()

hr.drop(columns=['Attrition_Num'], inplace=True)
```

## **Attrition by Compensation Factors**

```python
# 1. ensure 'Attrition_Num' exists to calculate the rate
if 'Attrition_Num' not in hr.columns:
    hr['Attrition_Num'] = hr['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# 2. set up the canvas for 2 subplots side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# boxplot of monthly income vs attrition
sns.boxplot(data=hr, x='Attrition', y='MonthlyIncome', palette='Set2', 
            showmeans=True, meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"black", "markersize":"8"}, 
            ax=axes[0])

axes[0].set_title('Monthly Income Distribution by Attrition', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Attrition Status', fontsize=10)
axes[0].set_ylabel('Monthly Income (USD)', fontsize=10)

# bar chart of stock option level vs attrition rate
attr_rate_stock = hr.groupby('StockOptionLevel')['Attrition_Num'].mean() * 100
sns.barplot(x=attr_rate_stock.index, y=attr_rate_stock.values, palette='Reds_r', ax=axes[1])

axes[1].set_title('Attrition Rate by Stock Option Level (%)', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Stock Option Level', fontsize=10)
axes[1].set_ylabel('Attrition Rate (%)', fontsize=10)

for p in axes[1].patches:
    axes[1].annotate(f'{p.get_height():.1f}%', 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='bottom', fontsize=10, xytext=(0, 5), textcoords='offset points')

# 3. adjust layout and display
plt.tight_layout()
plt.show()

hr.drop(columns=['Attrition_Num'], inplace=True)
```

## **Attritation by Tenure and Experience**

```python
# 1. list of tenure-related features to analyze
tenure_features = [
    'TotalWorkingYears', 'YearsAtCompany', 
    'YearsInCurrentRole', 'YearsSinceLastPromotion', 
    'YearsWithCurrManager'
]

# 2. loop through each feature to generate the visualizations
for feature in tenure_features:
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    # boxplot (distribution, median, and mean)
    sns.boxplot(data=hr, x='Attrition', y=feature, palette='Set2', 
                showmeans=True, 
                meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"black", "markersize":"8"}, 
                ax=axes[0])
    
    axes[0].set_title(f'Boxplot of {feature}', fontsize=12, fontweight='bold')
    axes[0].set_xlabel('Attrition Status', fontsize=10)
    axes[0].set_ylabel(f'{feature} (Years)', fontsize=10)
    
    # histogram (frequency/count distribution)
    sns.histplot(data=hr, x=feature, hue='Attrition', palette='Set2', 
                 multiple='dodge', bins=15, ax=axes[1])
    
    axes[1].set_title(f'Histogram of {feature}', fontsize=12, fontweight='bold')
    axes[1].set_xlabel(f'{feature} (Years)', fontsize=10)
    axes[1].set_ylabel('Employee Count', fontsize=10)
    
    # KDE plot (density/probability shape)
    sns.kdeplot(data=hr, x=feature, hue='Attrition', palette='Set2', 
                fill=True, common_norm=False, alpha=0.4, ax=axes[2])
    
    axes[2].set_title(f'KDE Distribution of {feature}', fontsize=12, fontweight='bold')
    axes[2].set_xlabel(f'{feature} (Years)', fontsize=10)
    axes[2].set_ylabel('Density', fontsize=10)
    
    # adjust layout to prevent overlapping text
    plt.tight_layout()
    plt.show()
```

# **Feature Engineering for Tableau**

```python
# attrition flag
hr['AttritionFlag'] = hr['Attrition'].map({
    'Yes': 1, 
    'No': 0})
```

```python
# age group
hr['AgeGroup'] = pd.cut(
    hr['Age'],
    bins=[0, 25, 35, 45, 55, 100],
    labels=['<=25', '26-35', '36-45', '46-55', '>55']
)
```

```python
# income group
hr['IncomeGroup'] = pd.cut(
    hr['MonthlyIncome'],
    bins=[0, 3000, 6000, 10000, 15000, hr['MonthlyIncome'].max()],
    labels=['Low', 'Lower-Mid', 'Mid', 'Upper-Mid', 'High'],
    include_lowest=True
)
```

```python
# tenure group
hr['TenureGroup'] = pd.cut(
    hr['YearsAtCompany'],
    bins=[-1, 1, 3, 5, 10, hr['YearsAtCompany'].max()],
    labels=['0-1 years', '2-3 years', '4-5 years', '6-10 years', '>10 years']
)
```

```python
# overtime risk label
hr['OverTimeRisk'] = hr['OverTime'].apply(
    lambda x: 'High Risk' if x == 'Yes' else 'Lower Risk'
)
```

```python
# promotion status
hr['PromotionStatus'] = hr['YearsSinceLastPromotion'].apply(
    lambda x: 'No promotion >=5 years' if x >= 5 else 'Recent promotion <5 years'
)
```

```python
# early career label
hr['CareerStage'] = hr['TotalWorkingYears'].apply(
    lambda x: 'Early Career' if x <= 5 else 'Experienced'
)
```

```python
# income per working year
hr['IncomePerWorkingYear'] = (
    hr['MonthlyIncome'] / (hr['TotalWorkingYears'] + 1)
)
```

```python
# role tenure ratio
hr['RoleTenureRatio'] = (
    hr['YearsInCurrentRole'] / (hr['YearsAtCompany'] + 1)
)
```
